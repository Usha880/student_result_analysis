```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())

```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop('Unnamed: 0',axis=1)
print(df.head())
```

```python
df.head()
```

```python
#Gender Distribution
plt.figure(figsize=(3,4))
plt.title("Gender Distribution")
ax=sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#From the above chart we analyzed that the number of females in the data is more than number of males
```

```python
#The impact of parent education on students marks

gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Relationship between parents's Education and Student's Score")
plt.show()
```

```python
#From the above map we have concluded that hte education of the parents have a good impact on their students education
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Relationship between parents's Marital Status and Student's Score")
plt.show()
```

```python
#From the above chart we have concluded that there is no negative impact of students score in regard with
#their parent marital status 
```

```python
#Distribution of ethnic groups

groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()

l=['groupA','groupB','groupC','groupD','groupE']
mlist=[groupA['EthnicGroup'], groupB['EthnicGroup'],groupC['EthnicGroup'] ,groupD['EthnicGroup'],
       groupE['EthnicGroup']]
print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
